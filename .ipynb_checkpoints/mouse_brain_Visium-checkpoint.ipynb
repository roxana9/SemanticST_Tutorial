```python
from sklearn import metrics
import torch
import os
import random
import numpy as np
from loading_batches import PrepareDataloader
from loading_batches import Dataloader
import scanpy as sc
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import torch.utils.data as data
import main
import warnings
warnings.filterwarnings("ignore")
```

<h3>2. Read data and import device!</h3>

```python
dataset="Mouse Brain"
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"You are using *{device}*")
BASE_PATH = Path('/media/rokny/DATA1/Roxana/Data/Adult_Mouse_Brain_Coronal_Section_1/')
spot_paths= Path(f'{BASE_PATH}')
```

<h3>3. Define the Spatial Transcriptomics (ST) Data Type and Run SemanticST </h3>
Available: ''Xenium', 'Visium', 'Stereo', Slide.
<br>
It is essential to specify the type of ST data, as different ST technologies require distinct preprocessing steps. Additionally, you have the option to select between mini-batch training for large datasets and full dataset training for smaller ones, ensuring efficient data processing and model performance.</h3>

```python
dtype = "Visium"  

result = main.main(spot_paths,device, dtype)  

if result[0] is not None:  
    Trainer = result[0]
    adata = result[1]
    config = result[2] 

    if config.use_mini_batch:  
        train_loader = result[3]
        test_loader = result[4]
        num_iter=result[5]
        model = Trainer(adata,train_loader, test_loader,num_iter,config)  
        adata=model.train() 

    else:  # If full dataset training
        model = Trainer(adata,config)  
        adata=model.train()  
```

```python
plt.rcParams["figure.figsize"] = (3, 3)
fig, ax = plt.subplots()
sc.pl.spatial(adata, img_key="hires",ax=ax)
```

<h3>4. Clustering </h3>
    - Louvain 

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=20, random_state=1) 
embedding = pca.fit_transform(adata.obsm['emb_decoder'].copy())
adata.obsm['emb_pca'] = embedding
sc.pp.neighbors(adata, use_rep='emb_pca')
sc.tl.umap(adata)
```

```python
sc.tl.louvain(adata, resolution=1.5)

#adata.obsm['spatial'][:, 1] = -1*adata.obsm['spatial'][:, 1]

adata.uns['colors']=['#aec7e8', '#9edae5', '#d62728', '#dbdb8d', '#ff9896',
                             '#8c564b', '#696969', '#778899', '#17becf', '#ffbb78',
                             '#e377c2', '#98df8a', '#aa40fc', '#c5b0d5', '#c49c94',
                             '#f7b6d2', '#279e68', '#b5bd61', '#ad494a', '#8c6d31',
                             '#1f77b4', '#ff7f0e']
plt.rcParams["figure.figsize"] = (4, 4)
fig, ax = plt.subplots()
sc.pl.spatial(adata,  color="louvain",title='SemanticST',palette=adata.uns['colors'], ax=ax,size=1.5)
ax.axis('off')
plt.show()
```

<h3>4. Clustering </h3>
    - Leiden

```python
sc.tl.leiden(adata, resolution=0.9)
plt.rcParams["figure.figsize"] = (4, 4)
fig, ax = plt.subplots()
sc.pl.spatial(adata,  color="leiden",title='SemanticST',palette=adata.uns['colors'], ax=ax,size=1.5)
ax.axis('off')
plt.show()
```
