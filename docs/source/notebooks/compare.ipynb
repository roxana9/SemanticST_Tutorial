```python
genes = ['IGFBP3', 'ACTA2']
import seaborn as sns

# Extract expression data for the two groups
def extract_gene_expression(adata, group_name, genes):
    # Ensure genes exist in the dataset
    filtered_genes = [gene for gene in genes if gene in adata.var_names]
    
    # Extract expression (convert to dense and reshape if necessary)
    expression_data = adata[:, filtered_genes].X
    if isinstance(expression_data, np.ndarray):
        expression_data = expression_data  # Already dense
    else:
        expression_data = expression_data.toarray()  # Convert sparse matrix to dense
    
    # Create DataFrame
    df = pd.DataFrame(expression_data, columns=filtered_genes)
    df['Group'] = group_name

    # Melt DataFrame to long format
    return df.melt(id_vars='Group', var_name='Gene', value_name='Expression')
# Extract for both groups
df_19 = extract_gene_expression(adata_MS19, 'Non-Responder', genes)
df_26 = extract_gene_expression(adata_NS26, 'Responder', genes)

# Combine the dataframes
combined_df = pd.concat([df_19, df_26], ignore_index=True)

# Set plotting style
sns.set(style="whitegrid")

# Boxplot: Compare CD68 and CD163 expression between responders and non-responders
plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Expression', hue='Gene', data=combined_df, palette='coolwarm', dodge=True, fliersize=0)

# Overlay scatterplot with jitter
sns.stripplot(x='Group', y='Expression', hue='Gene', data=combined_df, 
              dodge=True, jitter=True, palette='dark', size=3, alpha=0.7)

plt.title('CD24 and EpCAM Expression Levels in Responders and Non-Responders')
plt.ylabel('Expression Level')
plt.xlabel('Group')
plt.legend(title='Gene', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Violin Plot: Expression distribution
plt.figure(figsize=(8, 6))
sns.violinplot(x='Group', y='Expression', hue='Gene', data=combined_df, split=True, palette='coolwarm')
plt.title('MS4A1 and CD79A Expression Distribution in Responders and Non-Responders')
plt.ylabel('Expression Level')
plt.xlabel('Group')
plt.legend(title='Gene')
plt.tight_layout()
plt.show()

# Heatmap: Average expression per group
heatmap_data = combined_df.groupby(['Gene', 'Group'])['Expression'].mean().unstack()
plt.figure(figsize=(6, 4))
sns.heatmap(heatmap_data, annot=True, cmap='YlOrRd', fmt=".2f")
plt.title('Heatmap of Mean Expression Levels')
plt.ylabel('Gene')
plt.xlabel('Group')
plt.tight_layout()
plt.show()
```

```python
import scanpy as sc
import matplotlib.pyplot as plt
import numpy as np

# Define genes and corresponding color maps
genes = ["CD24", "EPCAM"]
gene_color_map = {
    "CD24": "Blues",
    "EPCAM": "Reds"
}
adata_NS26.obsm['spatial'][:, 1] = -1 * adata_NS26.obsm['spatial'][:, 1]
# Function to plot spatial expression with different colors
def plot_combined_spatial(adata, genes, gene_color_map, title):
    fig, ax = plt.subplots(figsize=(8, 8))
    
    for gene in genes:
        # Extract spatial coordinates and expression levels
        coords = adata.obsm["spatial"]
        expression = adata[:, gene].X
        
        # Ensure dense matrix if necessary
        if not isinstance(expression, np.ndarray):
            expression = expression.toarray().flatten()
        
        # Normalize expression for color mapping
        norm = plt.Normalize(vmin=expression.min(), vmax=expression.max())
        cmap = plt.get_cmap(gene_color_map[gene])
        
        # Scatter plot for the gene
        ax.scatter(
            coords[:, 0], coords[:, 1],
            c=cmap(norm(expression)),
            label=gene,
            s=1.5  # Adjust spot size as needed
        )
    
    ax.set_title(title)
    ax.legend()
    ax.axis("off")
    plt.tight_layout()
    plt.show()

# Combined spatial plot for responder
plot_combined_spatial(adata_NS26, genes, gene_color_map, title="Responder")

# Combined spatial plot for non-responder
#plot_combined_spatial(adata_non_responder, genes, gene_color_map, title="Non-Responder")

```

```python
# Select gene and extract spatial data
gene_of_interest = 'IGFBP3'
assert gene_of_interest in adata_MS19.var_names and gene_of_interest in adata_NS26.var_names, \
    f"{gene_of_interest} not found in dataset."

# Extract spatial coordinates and expression levels
def extract_spatial_expression(adata, gene, group_name):
    coords = adata.obsm['spatial']  # Assuming spatial coordinates are in 'spatial'
    expression = adata[:, gene].X.toarray().flatten()  # Convert to dense array if needed
    return pd.DataFrame({'x': coords[:, 0], 'y': coords[:, 1], 'Expression': expression, 'Group': group_name})

# Extract for both responders and non-responders
df_19_spatial = extract_spatial_expression(adata_MS19, gene_of_interest, 'Non-Responder')
df_26_spatial = extract_spatial_expression(adata_NS26, gene_of_interest, 'Responder')

# Combine data
spatial_combined_df = pd.concat([df_19_spatial, df_26_spatial], ignore_index=True)

# Plot spatial expression for responders and non-responders
sns.set(style="white")
g = sns.FacetGrid(spatial_combined_df, col='Group', height=6, aspect=1.2)
g.map_dataframe(sns.scatterplot, x='x', y='y', hue='Expression', size='Expression',
                palette='coolwarm', sizes=(20, 200), alpha=0.4, legend='brief')

# Add titles and adjust
g.set_axis_labels('X Coordinate', 'Y Coordinate')
g.set_titles('{col_name}')
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle(f'Spatial Expression of {gene_of_interest} in Responders and Non-Responders', fontsize=16)
g.add_legend(title='Expression Level')
plt.show()
```
